# First ML Project

#**Load data**

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv")
df
```

# **Data Preparation**

##**separation as x and y**

```python
y = df["logS"]
y
```

```python
x = df.drop("logS", axis=1)
x
```

##**Data Splitting**

```python

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

#**Model Building**

##**Linear Regression**

### **Training the model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

### **Applying the model to make prediction**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

### **Evaluate Model Performance**

```python
y_train #actual value
```

```python
y_lr_train_pred #predicted value
```

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print("LR TRAIN MSE", lr_train_mse)
print("LR TRAIN r2", lr_train_r2)
print("LR TEST MSE", lr_test_mse)
print("LR TEST r2", lr_test_r2)
```

```python
lr_results = pd.DataFrame(["Linear Regression" ,lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ["Method","TRAIN MSE", "TRAIN r2", "TEST MSE", "TEST r2"]
lr_results
```

## **Random Forest**

### **Training the model**

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

### **Applying the model to make prediction**

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

### **Evaluate Model Performance**

```python
rf_results = pd.DataFrame(["Random Forest Regressor" ,rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ["Method","TRAIN MSE", "TRAIN r2", "TEST MSE", "TEST r2"]
rf_results
```

## **Model Comparison**

```python
df_models = pd.concat([lr_results, rf_results] , axis=0).reset_index(drop=True)
df_models
```

#**Data Visualization**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
plt.scatter(x=y_train, y=y_lr_train_pred, c="#7CAE00", alpha=0.3)

z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_train, p(y_train), '#F8655D')
plt.xlabel("Experimental logS")
plt.ylabel("predicted logS")
plt.title("Linear Regression model")
plt.plot()
```
